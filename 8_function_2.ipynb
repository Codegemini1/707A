```python
def test7(*args):
    l = []
    for i in args:
        if i%2 == 0:
            l.append(i)
    return l
```

```python
test7(2,3,1,43,45,3,5,67,2,5,4,2)
```

```python
def test8(*args):
    l = []
    for i in args:
        if type(i) == int:
            l.append(i)
    return l
```

```python
test8(1.5,12,True,4,'Hello')
```

```python
for i in range(1,11):
    if i%2 == 0:
        print(i)
```

```python
[i for i in range(1,11) if i%2 ==0]
```
