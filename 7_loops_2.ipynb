# Break

```python
for i in range(1,6):
    if i == 3:
        break
    print(i)
```

```python
for i in 'python':
    if i == 'h':
        break
    print(i)
```

# Continue

```python
for i in range(1,6):
    if i == 3:
        continue
    print(i)
```

```python
for i in 'hello':
    if i == 'l':
        continue
    print(i)
```
