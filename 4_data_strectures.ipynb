```python
code = input('Enter your code for discount')
amount = int(input('Enter your total amount'))
if code == 'DS10':
    discount = (amount/100)*10
    print(f'Discount you will get {discount} and your final bill will be {amount - discount} ')
elif code == 'DS50':
    discount = (amount/100)*50
    print(f'Discount you will get {discount} and your final bill will be {amount - discount} ')
else:
    print(f'Your final amount will be {amount} ')
```
